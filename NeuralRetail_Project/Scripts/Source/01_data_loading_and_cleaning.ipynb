```python
df.dropna(inplace=True) #DRPONG NULL VALUE WE CAN USE FILLNA TO FILL THE NULL VALUES
```

```python
print(df.isnull().sum())
```

```python
print(df.duplicated().sum()) #TOTAL DUPLICATES
```

```python
df.drop_duplicates(inplace=True) #REMOVE DUPLICATES
```

```python
print(df.duplicated().sum())
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate']) # COVERT IT TO timestamp
```

```python
df['Quantity'] = df['Quantity'].astype(int) #convert it into int
df['Price'] = df['Price'].astype(float) #convert inti float
```

```python
df = df[df['Quantity'] > 0]

df = df[df['Price'] > 0]  #removes negative quantity
```

```python
df[df['Invoice'].astype(str).str.startswith('C')].head() 
```

```python
df = df[~df['Invoice'].astype(str).str.startswith('C')] #romve cancelled invoices
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)
```

```python
# cleaning notebook

df.to_csv(
    'cleaned_retail.csv',
    index=False
)
```
